```python
import os
import matplotlib.pyplot as plt

class_names = dataset.classes

counts = []
for class_name in class_names:
    class_path = os.path.join(dataset_path, class_name)
    counts.append(len(os.listdir(class_path)))

plt.figure(figsize=(8,5))
plt.bar(class_names, counts,color='maroon')
plt.title("Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.show()

```

```python
plt.figure(figsize=(6,6))
plt.pie(counts, labels=class_names, autopct='%1.1f%%')
plt.title("Class Distribution Percentage")
plt.show()
```

```python
from PIL import Image
import os

widths = []
heights = []

for i in range(1000):
    img_path, _ = dataset.samples[i]
    img = Image.open(img_path)
    widths.append(img.width)
    heights.append(img.height)

plt.figure(figsize=(8,5))
plt.scatter(widths, heights, alpha=0.3)
plt.title("Image Dimension Distribution (First 1000 Images)")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()

print("Average Width:", sum(widths)/len(widths))
print("Average Height:", sum(heights)/len(heights))
```
